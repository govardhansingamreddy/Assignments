```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
min_max_scaler = MinMaxScaler()
norm = min_max_scaler.fit_transform(data)
norm
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
plt.figure(figsize = (17,11))
dendrogram=sch.dendrogram(sch.linkage(norm, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=5, linkage = 'complete')
# Remove affinity parameter as it's not needed when linkage is 'complete'
# For linkage='ward', affinity is fixed to 'euclidean' in older versions
# If you need other distance metrics, consider upgrading scikit-learn
```

```python
y_hc = hc.fit_predict(norm)
y_hc
```

```python
hc.labels_
```

```python
data['h_clusterid'] = hc.labels_
```

```python
data2=data.drop(columns='h_clusterid')
data2.head()
```

```python
from sklearn.cluster import KMeans
scaler = StandardScaler().fit_transform(data2)
scaler
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaler)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
wcss
```

```python
clusters = KMeans(6, random_state=42)
clusters.fit(scaler)
clusters.labels_
```

```python
clusters.fit_predict(scaler)
```

```python
clusters.inertia_
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(data2, clusters.labels_)
```

```python
c2=KMeans(2, random_state=41)
c2.fit(scaler)
```

```python
silhouette_score(data2, c2.labels_)
```
